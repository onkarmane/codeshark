# Indexing of arrays

1. Commas separate axes of an array.
2. Colons mean "through". For example, x[0:4] means the first 5 rows (rows 0 through 4) of x.
3. Negative numbers mean "from the end of the array." For example, x[-1] means the last row of x.
4. Blanks before or after colons means "the rest of". For example, x[3:] means the rest of the rows in x after row 3.
   Similarly, x[:3] means all the rows up to row 3. x[:] means all rows of x.

```python
import numpy as np
```

```python
arr = np.arange(1,17)
```

```python
arr
```

```python
arr = arr.reshape(4,4)
```

```python
arr
```

```python
arr.shape
```

```python
# Output 15 ,16
arr[3,2:]
```

```python
# Output last column
arr[:,3]
```

```python
# Output 9,10,13,14
arr[2:,:2]
```

# vstack and hstack


```python
a = np.ones((3, 3))
print(a)
b= np.arange((3))
print(b)

```

```python
np.vstack( (a,b ) )
```

```python
#hstack
c = np.ones((4, 4))
print(c)
d= np.zeros((4,))
print(d)
```

```python
d = d.reshape(4,1)
```

```python
d
```

```python
np.hstack((c,d))
```

# .loc[rows , columns] >> Name of rows and columns
All indexes included

```python
import pandas as pd

# making data frame from csv file
data = pd.read_csv("nba.csv", index_col ="Name")

data
```

```python
# retrieving all rows and some columns by loc method
first = data.loc[:, ["Team", "Number", "Position"]]
print(first)
```

```python
# retrieving some rows and some columns by loc method
first = data.loc['Avery Bradley':'Jonas Jerebko', ["Team", "Number", "Position"]]
print(first)
```

# Indexing a DataFrame using .iloc[  ] >> Number of row and column


```python
import pandas as pd

# making data frame from csv file
data = pd.read_csv("nba.csv", index_col ="Name")

data
```

```python

# retrieving all rows and some columns by iloc method
row2 = data.iloc [:, [1, 2]]
print(row2)
```

```python
#Last index is excluded
row2 = data.iloc [1:4, [1, 2]]
print(row2)
```

# Conditional Indexing

```python
# importing pandas
import pandas as pd

record = {

'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ],
'Age': [21, 19, 20, 18, 17, 21],
'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],
'Percentage': [88, 92, 95, 70, 65, 78] }

# create a dataframe
dataframe = pd.DataFrame(record, columns = ['Name', 'Age', 'Stream', 'Percentage'])

print("Given Dataframe :\n", dataframe)


```

```python
# selecting rows based on condition
rslt_df = dataframe[dataframe['Percentage'] > 80]

print('\nResult dataframe :\n', rslt_df)
```

```python
# selecting rows based on condition
rslt_df = dataframe.loc[dataframe['Percentage'] != 95]
  
print('\nResult dataframe :\n', rslt_df)
```

# isin()

```python
# importing pandas
import pandas as pd

record = {
'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ],
'Age': [21, 19, 20, 18, 17, 21],
'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],
'Percentage': [88, 92, 95, 70, 65, 78]}

# create a dataframe
dataframe = pd.DataFrame(record, columns = ['Name', 'Age', 'Stream', 'Percentage'])

print("Given Dataframe :\n", dataframe)

options = ['Math', 'Commerce']

```

```python
# selecting rows based on condition
rslt_df = dataframe[dataframe['Stream'].isin(options)]

print('\nResult dataframe :\n', rslt_df)
```

```python
# importing pandas
import pandas as pd

record = {
'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ],
'Age': [21, 19, 20, 18, 17, 21],
'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],
'Percentage': [88, 92, 95, 70, 65, 78]}

# create a dataframe
dataframe = pd.DataFrame(record, columns = ['Name', 'Age', 'Stream', 'Percentage'])

print("Given Dataframe :\n", dataframe)

options = ['Math', 'Science']


```

```python
# selecting rows based on condition
rslt_df = dataframe[(dataframe['Age'] == 21) &
		dataframe['Stream'].isin(options)]

print('\nResult dataframe :\n', rslt_df)
```

# Suppose I want to make a choice between 2 options then use if-else

```python
l = [45,38,78,45,12,30,89]
```

# More examples on for loop with if-else

```python
for item in range(10):
    if item < 5:
        print("Number is less than 5")
    elif item >= 5 and item <= 7:
        print("Number is between 5 and 7")
    else:
        print("Number is greator than 7")
```

# This looks not very great....instead of "Number", I want actual number.

```python
for item in range(10):
    if item < 5:
        print(f"{item} is less than 5")
    elif item >= 5 and item <= 7:
        print(f"{item} is between 5 and 7")
    else:
        print(f"{item} is greator than 7")
```
